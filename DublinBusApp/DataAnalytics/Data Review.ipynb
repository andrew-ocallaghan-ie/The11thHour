## Definitions and Terminology:


### Signature:
<p align="justify">For the intents and purposes of this review a signature is a pattern of blocks/bus stops which a bus visits. For each unique Journey_Pattern_ID it is expected that the signitures for different journeys are identical (or atleast closely resemble one another).</p>

<p align="justify"> The signatures for LineIds and Journey_Pattern_IDs do not correspond. After investigating this phenomena it was determined that Journey_Pattern_IDs are the most trustworthy and convinient soure for LineID and direction infromation.
</p>

## Columns Info:
This section concerns columns and their ambiguities and limitations

   ### Timestamp: 
   
           - A large int representing the time (to microsecond precision) since UNIX Epoch.
   
   
   ### Line_ID:
   
           - Identifies the "line"/route of the bus. Many of these appear to be incorrect; 
       
           - Using signitures we can differentiate between different paths buses have taken. 
               - Many of these signitures do not correspond to the Line that we expect from the LineID.
       
           - The LineID infromation can be re-derived form the Journey_Pattern_ID infromation.
   
   
   ### Direction:
   
           - It was expected that this would signify a sort of "inbound/outbound" nature for a particular Journey Pattern.
               - It is infact a constant column. 
   
           - This can be re-derived using info from Journey_Pattern_ID
  
  
   ### Journey_Pattern_ID: (IMPORTANT)
   
   
           - It is expected that this uniqely identifies (or implies) a certain  signiture or pattern.
               - The path which a bus has taken.
       
           - We believe a Vehicle_Pattern_ID describes a set of Block_IDs (which in turn describe a set of Stop_IDs).
   
   
           - The First FOUR characters of this give a "stronger" indication of the route than Line_ID.
               - e.g 029Axxxx implies that the route is the 29A (often conflict with the Line_ID)
       
           - The fifth character seems to carry some form of "directional" infromation. 
               - xxxx1--- and xxxx0--- will have a similar number of occurances. 
                   - (As expected on an inbound/outbound route behaviour.)
       
           -The remaining three characters we believe are related to the block_id. 
               - The current working hypothesis is that they describe the departure/destination block
                   - (i.e. the src of the journey or its final destination)


   ### Time_Frame:
      
           - The similar to the date in the timestamp. 
               - it concerns a period of time begining at 6am and ending and 3am (21 hours later).



   ### Vehicle_Journey_ID:
   
           - An ID for a given journey (a single run from one end of the line to another)
           - 30% nulls?
       
 
 
 
   ### Operator
   
           - The Terminus/Bus Station where the buses are kept and maintained.
   
 
 
   ### Congestion:
   
          - a binary column depicting if there is traffic or not?
   
 
 
   ### Latitude / Logitude:
   
           - Columns giving geographic coordinates of the bus at a given time (timestamp)
           - Can these be used to extract bus stop info
           - 
   
 
 
   ### Delay:
   
           - Given in seconds. 
           - A positive delay indicates a bus running behind schedule.
           - A negative delay indicates a bus running ahead of schedule.
           - A delay of zero would be a bus exatly on time.
           - From investigation, Delay is relative to schedule. Some results are in conflict
   
 
 
   ### Block_ID:
   
            - A section id of the journey pattern. Describes a set of Stop_IDs
            - Can different routes overlap on Blocks?
   
 
 
   ### Vehicle_ID:
   
           - Uniquly identifies a physical bus.
   
 
 
   ### Stop_ID:
   
            - Uniquely identifies a bus stop. E.g. stop 570 is the physical bus stop at the top of Thormanby Woods.
   


  ### At_stop:
  
          - Binary column which signifies if a bus is at a bus stop or not.
          - Does it pick up when a bus passes a bus stop, or only when it has stopped at one in that journey
          - Or only when a timestamp signal is sent out, while it is at the stop (if those two overlap exactly)
